# Movie Review Analyzer

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import Pipeline
```

```python
reviews = [
    "The movie was amazing with great acting and story",
    "I really hated this movie it was boring and slow",
    "Fantastic film with excellent visuals and music",
    "Worst movie ever waste of time",
    "It was an average movie not too good not too bad"
]

sentiments = ["positive", "negative", "positive", "negative", "neutral"]
```

```python
df = pd.DataFrame({
    "review": reviews,
    "sentiment": sentiments,
})

df
```

```python
model = Pipeline([
    ('tfidf', TfidfVectorizer(stop_words='english')),
    ('classifier', MultinomialNB())
])
```

```python
test_reviews = [
    "This movie was fantastic and exciting",
    "I did not like the film it was dull",
    "It was okay not the best",
    "Absolutely loved the performance",
    "Terrible storyline and poor acting"
]

predictions = model.predict(test_reviews)

for review, sentiment in zip(test_reviews, predictions):
    print(f"Review: {review}")
    print(f"Predicted Sentiment: {sentiment}")
    print("-" * 50)
```
