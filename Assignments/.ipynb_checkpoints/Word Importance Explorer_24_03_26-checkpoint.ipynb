```python
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd
```

```python
documents = [
    "Artificial Intelligence is transforming the world with automation",
    "Machine learning and artificial intelligence are closely related fields",
    "Cloud computing enables scalable and flexible IT solutions",
    "Cyber security protects systems from digital attacks and threats",
    "DevOps combines development and operations for faster delivery"
]
```

```python
# TF-IDF 
vectorizer = TfidfVectorizer(stop_words='english')
```

```python
# Transform data
tfidf_matrix = vectorizer.fit_transform(documents)
```

```python
# Convert to DataFrame
df = pd.DataFrame(tfidf_matrix.toarray(), columns=vectorizer.get_feature_names_out())
```

```python
# Get top words per document
top_words = {}

for i in range(len(documents)):
    top = df.iloc[i].sort_values(ascending=False).head(3)
    top_words[f"Doc{i+1}"] = top.index.tolist()

print(top_words)
```
