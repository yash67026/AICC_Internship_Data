```python
import pandas as pd
```

```python
resume_prompts = {
    "Weak": "Write a resume",
    
    "Strong": """Create a professional resume for a Computer Science student 
    skilled in Python, Machine Learning, and AWS. Include sections: Summary, 
    Skills, Projects, Education, and Certifications. Keep it concise and ATS-friendly."""
}
```

```python
business_prompts = {
    "Weak": "Give me a business idea",
    
    "Strong": """Suggest a scalable tech-based business idea for India 
    targeting college students. Include problem, solution, revenue model, 
    and why it will succeed in the current market."""
}
```

```python
study_prompts = {
    "Weak": "Make a study plan",
    
    "Strong": """Create a 30-day study plan for learning DevOps including 
    Linux, Docker, Kubernetes, and AWS basics. Allocate daily tasks 
    with practical exercises and revision strategy."""
}
```

```python
data = [
    ["Resume", resume_prompts["Weak"], resume_prompts["Strong"]],
    ["Business Idea", business_prompts["Weak"], business_prompts["Strong"]],
    ["Study Plan", study_prompts["Weak"], study_prompts["Strong"]],
]

df = pd.DataFrame(data, columns=["Category", "Weak Prompt", "Strong Prompt"])
df
```

```python
df["Weak Output"] = ["Output1", "Output2", "Output3"]
df["Strong Output"] = ["Output1", "Output2", "Output3"]
```

```python
df["Observation"] = [
    "Weak prompt gives generic resume, strong prompt gives structured and relevant resume",
    "Weak prompt gives random idea, strong prompt gives detailed and practical idea",
    "Weak prompt gives vague plan, strong prompt gives step-by-step plan"
]

df
```
