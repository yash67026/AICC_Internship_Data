```python
import spacy

# Load medium English model
nlp = spacy.load("en_core_web_md")
```

```python
word_pairs = [
    ("car", "automobile"),
    ("happy", "joyful"),
    ("teacher", "instructor"),
    ("computer", "laptop"),
    ("river", "ocean")
]
```

```python
results = []

for w1, w2 in word_pairs:
    token1 = nlp(w1)
    token2 = nlp(w2)
    
    similarity = token1.similarity(token2)
    results.append((w1, w2, similarity))

for w1, w2, score in results:
    print(f"{w1} - {w2} : {score:.2f}")
```
